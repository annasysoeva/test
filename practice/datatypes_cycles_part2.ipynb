```python
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]

dicts = {}

for req in queries:
    word = req.split()

    if len(word) in dicts.keys():
        dicts[len(word)] += 1
    else:
        dicts.update({
            len(word): 1
        })

for k, v in dicts.items():
    perc = round((v / len(queries)) * 100, 2)
    print('Поисковых запросов из', {k}, 'слов(а): ', {perc}, '%')
```
