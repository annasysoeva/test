```python
class emp:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
        
    def up(self):
        self.salary += 100
        
    def printt(self):
        print('Сотрудник', self.name, ', зарплата', self.salary)
```

```python
ivan = emp(name = 'Иван', salary = 100)
```

```python
ivan.printt()
ivan.up()
ivan.printt()
```
