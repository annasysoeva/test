```python
def minus(number):
    result = number - 2
    return result

print(minus(10))
```

```python
def plus(number, number_2=2):
    result = number - number_2
    return result

plus(10)
```

```python
def poly(word):
    reverse = word[::-1]
    return word.lower() == reverse.lower()

poly('Радар')
```

```python
dict_1 = {'flat_1': 10500, 'flat_2': 11000}
dict_2 = {'flat_3': 15000}
dict_3 = {'flat_4': 6500, 'flat_5': 7000, 'flat_6': 6000}

def avg_price(*neigh):
    total_price = 0
    ap_count = 0
    
    for neighs in neigh:
        total_price += sum(neighs.values())
        ap_count += len(neighs.values())
    return total_price / ap_count

avg_price(dict_3, dict_1)
```

```python
prices = [[100, 200, 400, 600], [200, 500], [100, 200, 100, 100], [800, 900]]

def avg_pr(lst):
    return [sum(x) / len(x) for x in lst]

avg_pr(prices)
```

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
def ds():
    del_shelf = input('Введите номер полки: ')
    
    for k, v in directories.items():
        if del_shelf not in directories:
            return(print(f"Такой полки не существует. Текущий перечень полок: {list(directories.keys())}"))
        elif del_shelf in directories and directories.get(del_shelf) == []:
            del(directories[del_shelf])
            return(print(f"Полка удалена. Текущий перечень полок: {list(directories.keys())}"))
        else:
            return(print(f"На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {list(directories.keys())}"))
```

```python
def ads():
    new_shelf = input('Введите номер полки: ')
    
    for dicts in directories.items():
        if new_shelf in dicts[0]:
            return(print(f"Такая полка уже существует. Текущий перечень полок: {list(directories.keys())}"))
    else:
        directories[new_shelf] = []
        return(print(f"Полка добавлена. Текущий перечень полок: {list(directories.keys())}"))  
```

```python
def l():
    for dicts in documents:
        file = list(filter(lambda x: dicts['number'] in x[1], directories.items()))[0]
        dicts['directory'] = file[0]
        print(dicts)
```

```python
def s():
    num = input("Введите номер документа:")
    
    for k, v in directories.items():
        if num in v:
            return(print('Документ хранится на полке: ', k))
        else:
            print('Документ не найден в базе')
```

```python
def p():
    num = input("Введите номер документа:")
    
    for dicts in documents:
        if dicts['number'] == num:
            print('Владелец документа: ', dicts["name"])
        else:
            print("Документ не найден в базе")
```

```python
prompt = 'Введите комманду: '
message = ""
while message != 'q':
    message = input(prompt)
    print(message)
    
    if message == 'p':
        print(p())
    elif message == 's':
        print(s())
    elif message == 'l':
        print(l())
    elif message == 'ads':
        print(ads())
    elif message == 'ds':
        print(ds())  
```
