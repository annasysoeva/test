# ДЗ 4. Выборка групп и разделение на А/Б. t-tests

```python
import pandas as pd
from scipy import stats as st
```

```python
data = pd.read_csv('Netology_A_B-testing_HW_1.csv')
```

## 1. Android => более высокие юзерские отмены

H0: у пользователей Android более высокие юзерские отмены
H1: у пользователей Android юзерские отмены меньше или равны тем, что у пользователей iOS

```python
# распределение по os
android = data.loc[data['os'] == 'Android']
ios = data.loc[data['os'] == 'iOS']
```

```python
# распределение по отменам 
android_c = android.loc[android['user_cancelled'] == 1]
ios_c = ios.loc[ios['user_cancelled'] == 1]
```

```python
alpha = 0.05

result = st.ttest_ind(android_c['user_cancelled'], ios_c['user_cancelled'], equal_var=False)
print(result)
if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
```

### Вывод: не отвергаем нулевую гипотезу. У пользователей Android действительно могут быть более высокие юзерские отмены

## 2. Заказ из центра => более низкие юзерские отмены

H0: при заказе из центра юзерские отмены выше, чем при заказе из других частей города  
H1: при заказе из центра юзерские отмены ниже или такие же, как при заказе из других частей города

```python
# распределение по заказам
centre = data.loc[data['city_center_order'] == 1]
not_centre = data.loc[data['city_center_order'] == 0]
```

```python
# распределение по отменам 
centre_c = centre.loc[centre['user_cancelled'] == 1]
not_centre_c = not_centre.loc[not_centre['user_cancelled'] == 1]
```

```python
alpha = 0.05

result = st.ttest_ind(not_centre_c['user_cancelled'], centre_c['user_cancelled'], equal_var=False)
print(result)
if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
```

### Вывод: не отвергаем нулевую гипотезу. При заказе из центра юзерские отмены выше
