# ДЗ 3. Основы распределения и статистики

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Netology_A_B-testing_HW_1.csv')
```

# 1. Рассчет метрик

№1. Количество заказов

```python
# среднее
data.loc[data['order_made'] == 1].mean()
```

```python
# медиана
data.loc[data['order_made'] == 1].median()
```

№ 2. Количество отмен

```python
# среднее
data.loc[data['user_cancelled'] == 1].mean()
```

```python
# медиана
data.loc[data['user_cancelled'] == 1].median()
```

№ 3. Количество поездок с surge

```python
# среднее
data.loc[data['surge'] == 'surge'].mean()
```

```python
# медиана
data.loc[data['surge'] == 'surge'].median()
```

№ 4. Распределение по возрасту

```python
# среднее
data['age'].mean()
```

```python
# медиана
data['age'].median()
```

```python
# персентиль
a = data['age']
print(np.percentile(a,95))
```

```python
# частотный график
a = data['age']
sns_plot = sns.distplot(a)
fig = sns_plot.get_figure()
```

```python
# boxplot
data.age.plot(kind = 'box')
```

№ 5. Распределение по дистанции поездки

```python
# среднее
data['distance'].mean()
```

```python
# медиана
data['distance'].median()
```

```python
# частотный график
a = data['distance']
sns_plot = sns.distplot(a)
fig = sns_plot.get_figure()
```

```python
# boxplot
data.distance.plot(kind = 'box')
```
